```python
import csv
from collections import Counter
```

```python
%pylab inline
```

```python
import matplotlib.pyplot as plt
```

```python
with open('/docgraph/ha1c_links.csv', "r") as f:
    reader = csv.DictReader(f)
    
    states = [record['state'] for record in reader]
    
    c = Counter(states)
    
    topstates, topcounts = zip(*c.most_common())
```

```python
#plt.bar(range(len(topcounts)), topcounts);
#plt.xticks(range(len(topcounts)), topstates, rotation=90);
plt.bar(range(10), topcounts[:10], align='center');
plt.xticks(range(10), topstates[:10]) #, verticalalignment= 'bottom', horizontalalignment='right');
```
